# Pandas

## Processamento de dados I

````python
import pandas as pd
````

````python
df = pd.read_csv('alunos.csv', sep=';')
````

````python
df
````

````python
type(df)
````

Podemos acessar os valores nas colunas pelos seus valores:

````python
df['Nome']
````

Acessando os valores da coluna dessa forma teremos um objeto do tipo 'Series':

````python
type(df['Nome'])
````

As séries podem ser mais importantes do que um dataframe porque um dataframe nada mais é do que um conjunto de séries.

Se quisermos obter uma lista com os valores propriamente ditos da coluna, usamos o método tolist():

````python
df['Nome'].tolist()
````

Uma lista de colunas permite selecionar várias colunas simultaneamente. Quando é feita essa seleção, o pandas retorna uma subtabela, que é uma cópia da tabela com somente as colunas selecionadas.

````python
df[['Nome', 'Frequencia']]
````

Para selecionar uma linha, utilizar o método .loc com o índice da linha desejada.

````python
df.loc[2]
````

O método .loc aceita dois argumentos: o índice de linha e o de coluna. Para alterar o valor da nota da Prova 2 de Carlos Vernes:

````python
df.loc[2, 'Prova_2'] = 8.5
````

Para visualizar o resultado da alteração:

````python
df.loc[2]
````

Para criar uma nova coluna, com todas as linhas preenchidas com o mesmo valor: 

````python
df['Nova_Coluna'] = 'Nova'
````

````python
df
````

Para criar uma nova coluna, com uma lista aleatória de notas de uma suposta Prova 5:

````python
import random
````

````python
df['Prova_5'] = [random.randint(0,10) for _ in range(df.shape[0])]
````

````python
df
````

Para descobrir o número de linhas e colunas:

````python
df.shape
````

Como se trata de uma tupla, é possível utilizar os índices para determinar a quantidade de linhas e colunas do dataframe:

````python
df.shape[0]
````

````python
df.shape[1]
````

Também é possível criar uma nova linha:

````python
df.loc[10, 'RA'] = 110237
df.loc[10, 'Nome'] = 'Hans Solo'
df.loc[10, 'Frequencia'] = 75
df.loc[10, 'Prova_1'] = 10
````

````python
df
````

Para **apagar** uma coluna:

````python
df.drop(columns=['Nova_Coluna'])
````

Quando é feito esse drop, o pandas retorna uma subtabela, que é uma cópia da tabela com somente as colunas dropadas.

````python
df
````

Para alterar o próprio dataframe:

````python
# df = df.drop(columns=['Nova_Coluna']) ou então,
df.drop(columns=['Nova_Coluna'], inplace=True)
````

````python
df
````

Para **apagar** uma linha:

````python
df.drop(index=[10], inplace=True)
````

````python
df
````

Podemos fazer operações entre os valores das colunas, e com isso criar novas colunas!

````python
df['Media'] = (df['Prova_1'] + df['Prova_2'] + df['Prova_3'] + df['Prova_4'] + df['Prova_5']) / 5
````

````python
df
````

Para obter informações a respeito dos tipos de dados do dataframe:

````python
df.info()
````

Existem algumas funções prontas:

````python
df.columns
````

````python
df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4', 'Prova_5']]
````

Para calcular a **média das colunas**:

````python
df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4', 'Prova_5']].mean() # Por padrão axis = 0.
````

Para calcular a **média das linhas** (axis=1)

````python
df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4', 'Prova_5']].mean(axis=1)
````

Uma **maneira programática** de realizar a média das linhas (depende da estrutura do nome da coluna):

````python
df['Media_2'] = df[[col for col in df.columns if "Prova_" in col]].mean(axis=1)
````

````python
df
````

Podemos fazer **filtros** muito facilmente. Basta explicitarmos condições sobre os valores das colunas e utilizar isso como indexador do dataframe!

````python
# mascara = df['Media'] > 8
# df[mascara]

df[df['Media'] > 8]
````

Se quisermos fazer filtros mais complexos (filtros compostos, em mais de uma coluna), podemos fazer **conjunções entre filtros**, utilizando os **operadores lógicos de conjunção**.

- & corresponde ao "and";
- | corresponde ao "or";
- ~ corresponde ao "not".

````python
df[(df['Media'] > 7) & (df['Frequencia'] < 20)]
````

Uma outra forma de fazer filtros é utilizando o método ```query()```, que permite que passemos uma expressão lógica na forma de uma string:

````python
df.query('Media > 7 and Frequencia < 20')
````

Para salvar um dataframe de volta para CSV:

````python
df.to_csv('novas_notas.csv', sep=';', index=False) # index=False é útil para não salvar os índices das linhas
````
